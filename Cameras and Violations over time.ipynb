```python
import pandas as pd
df = pd.read_csv('Speed_Camera_Violations.csv')
df
```

```python
df['date'] = pd.to_datetime(df['VIOLATION DATE'])
```

```python
df
```

```python
grouped_date = df.groupby('date')
```

```python
grouped_date.count()
```

```python
Date = []
Violations = []
for date, df_date in grouped_date:
    violations_per_day = df_date.sum()['VIOLATIONS']
    Date.append(date)
    Violations.append(violations_per_day)
    print "The total number of violations on %s are %d" %(date,violations_per_day)
```

```python
Date = []
Cameras = []
for date, df_date in grouped_date:
    total_cameras = df_date.count()['CAMERA ID']
    Date.append(date)
    Cameras.append(total_cameras)
    print "The total number of cameras on %s are %d" %(date,total_cameras)
```

```python

from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')
%matplotlib inline
plt.figure(figsize=(20,10))
plt.plot(Date, Cameras)

plt.title('City of Chicago - Number of cameras plotted against time')
plt.xlabel('Time')
plt.ylabel('Number of Active Cameras')

```

```python
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')
%matplotlib inline
plt.figure(figsize=(20,10))
plt.plot(Date, Violations)

plt.title('City of Chicago - Number of violations plotted against time')
plt.xlabel('Time')
plt.ylabel('Number of Violations observed')
```
