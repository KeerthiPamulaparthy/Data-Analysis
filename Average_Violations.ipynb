```python
import pandas as pd
df = pd.read_csv('Speed_Camera_Violations.csv')
df
```

```python
from datetime import date
num_days = df['date'][69340] - df['date'][0]
print num_days
```

```python
grouped_cam = df.groupby('CAMERA ID')
```

```python
Camera_ID = []
Average_violations = []
total_ave = 0
total = 0
for cam, cam_df in grouped_cam:
    n_violations = cam_df.sum()['VIOLATIONS'] #sum of all the violations, grouped for each camera
    n_count = cam_df.count()['VIOLATIONS'] #number of days the violations have reported
    ave = n_violations / n_count #average for each camera
    total_ave = ave + total_ave #sum of averages
    total = 1 + total #number of cameras
    Camera_ID.append(cam)
    Average_violations.append(ave)
    print '%s : %d' %(cam, ave)
Ave_cam_day = total_ave/ total
print 'sum of averages = %d'%total_ave
print 'Average per camera per day = %d' %(Ave_cam_day)
```
