```python
import pandas as pd
df = pd.read_csv('Speed_Camera_Violations.csv')
df
```

```python
df['date'] = pd.to_datetime(df['VIOLATION DATE'])
```

```python
df
```

```python
df['weekday'] = df['date'].dt.dayofweek
```

```python
df
```

```python
days = {0:'Monday', 1:'Tuesday', 2:'Wednesday', 3:'Thursday', 4:'Friday', 5:'Saturday', 6:'Sunday'}
```

```python
df['weekday'] = df['weekday'].apply(lambda x: days[x])
```

```python
df['weekday']
```

```python
grouped_weekday = df.groupby('weekday')
```

```python
grouped_weekday
```

```python
df_Monday= grouped_weekday.get_group('Monday')
```

```python
df_Monday.sum()['VIOLATIONS']
```

```python
for day, df_weekday in grouped_weekday:
    total_violations = df_weekday.sum()['VIOLATIONS']
    
    print "The total citations issued on %s are %d" %(day, total_violations)
    
```
